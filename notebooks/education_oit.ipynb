```python
import pandas as pd
```

```python
education_csv = '../data/raw/educacion.csv'
education = pd.read_csv(education_csv)
education
```

```python
label = education['indicator.label'].unique()[0].lower().replace('/','-').replace(' ', '_')
label
```

```python
# Select Venezuela data
education_vnzla = education[education['ref_area.label'].str.contains('Venezuela')]
education_vnzla = education_vnzla[['sex.label', 'classif1.label', 'classif2.label', 'time', 'obs_value']]

# Clean column names
education_vnzla.columns = ['sex', 'age_group', 'education_level', 'time', 'obs_values']

# Clean sex and age_group columns
education_vnzla['sex'] = education_vnzla['sex'].str.replace('Sex: ', '')
education_vnzla['age_group'] = education_vnzla['age_group'].str.replace('Age (Youth bands): ', '')
education_vnzla['education_level'] = education_vnzla['education_level'].str.replace('Education (Aggregate levels): ', '')

# Add total columns
education_vnzla['is_sex_group_total'] = education_vnzla['sex'] == 'Total'
education_vnzla['is_education_level_group_total'] = education_vnzla['education_level'] == 'Total'
education_vnzla['is_overall_total'] = (education_vnzla['sex'] == 'Total') & (education_vnzla['education_level'] == 'Total')

# Reset the index
education_vnzla.reset_index(drop=True, inplace=True)

# Save the cleaned DataFrame to a new CSV file
education_vnzla.to_csv(f'../data/processed/cleanned_{label}.csv', index=False)
```
